### Explanation of the export code

`summons_sample` is the dataframe we want to export

`.to_csv()` saves our dataframe to a .csv file. You are baseically exporting your python data out into a format that Excel, Google Sheets can read.

`index=False`
By default, on export pandas includes the DataFrame's index – the row numbers (0, 1, 2, 3...). We NEVER
need it 99.99 percent of the time. `index=False` simply tells pandas to leave that column out on export.

`encoding="utf-8"`
Tells Python which character encoding to use when writing the file. `UTF-8` is the most universal standard — it handles English plus accented characters, symbols, non-Latin scripts, etc. without garbling them. If your data has names like "García" or "Müller" and you skip this, you might get weird characters or errors depending on your system.

### Prompt: 

Provide the pandas code to calculate which boroughs had the most summons (in descending order).


````python
# 1. Which borough had the most summons? 

# Replace string '(null)' placeholders with real NaN so they're excluded cleanly
nypd_summons['BORO'] = nypd_summons['BORO'].replace('(null)', pd.NA)

# Count summons per borough, sort descending, and drop any missing borough values
borough_counts = (
    nypd_summons['BORO']
    .value_counts()              # Count occurrences of each borough; sorts descending by default
    .reset_index()               # Convert the Series into a two-column DataFrame
    .rename(columns={            # Rename columns from pandas defaults to readable names
        'BORO': 'borough',
        'count': 'summons_count'
    })
)

borough_counts
````

````python
# 2. What were the top 10 most common offenses?

### Prompt: Provide the pandas code to calculate what the top 10 most common offenses were.

import pandas as pd

df = pd.read_csv("summons-15-25.csv")

top_10_offenses = (
    df["OFFENSE_DESCRIPTION"]
    .value_counts()
    .head(10)
    .rename_axis("Offense")
    .reset_index(name="Count")
)

print(top_10_offenses)
````

````python
# 3. What are the total number of summons broken down by year?

### Prompt: Provide the pandas code to calculate the total number of summons broken down by year

import pandas as pd

df = pd.read_csv("summons-15-25.csv")

summons_by_year = (
    df.groupby("YEAR")["SUMMONS_KEY"]
    .count()
    .reset_index(name="Total_Summons")
)

print(summons_by_year)
````

````python
# 4. What were some of the most unusual summons?

### Prompt: Provide the panadas code to calculate the most unusual summons, defined as the rarest offense types

import pandas as pd

df = pd.read_csv("summons-15-25.csv")

rarest_offenses = (
    df["OFFENSE_DESCRIPTION"]
    .value_counts()
    .tail(10)
    .rename_axis("Offense")
    .reset_index(name="Count")
    .sort_values("Count")
)

print(rarest_offenses)
````

````python
### Prompt: Provide the panadas code to calculate the most unusual summons, defined as by demographic and offending summons

import pandas as pd

df = pd.read_csv("summons-15-25.csv")

# Define demographic and offense fields
demographic_fields = ["AGE_GROUP", "SEX", "RACE"]
offense_fields = ["OFFENSE_DESCRIPTION", "SUMMONS_CATEGORY_TYPE"]

key_fields = demographic_fields + offense_fields

# Calculate frequency of each value per field
for field in key_fields:
    freq = df[field].value_counts(normalize=True)
    df[f"{field}_freq"] = df[field].map(freq)

# Compute separate rarity scores for demographics and offense
demo_freq_cols = [f"{field}_freq" for field in demographic_fields]
offense_freq_cols = [f"{field}_freq" for field in offense_fields]

df["demographic_rarity"] = df[demo_freq_cols].prod(axis=1)
df["offense_rarity"] = df[offense_freq_cols].prod(axis=1)

# Combined rarity score (equal weighting of both dimensions)
df["combined_rarity"] = df["demographic_rarity"] * df["offense_rarity"]

# Surface the most unusual summons
most_unusual = (
    df.nsmallest(10, "combined_rarity")
    [["SUMMONS_KEY", "SUMMONS_DATE"]
     + demographic_fields
     + offense_fields
     + ["demographic_rarity", "offense_rarity", "combined_rarity"]]
)

print(most_unusual)
````

````python
# 5. What is the count of summons by gender?

### Provide the pandas code for the count of summons by gender
# Replace string '(null)' placeholders with real NaN so they're excluded cleanly

import pandas as pd

df = pd.read_csv("summons-15-25.csv")

summons_by_gender = (
    df["SEX"]
    .value_counts()
    .rename_axis("Gender")
    .reset_index(name="Count")
)

print(summons_by_gender)
````

````python
# 6. How many summons were issued by race?

### Provide the pandas code to find summons issued by raced (ordered)

import pandas as pd
import numpy as np

df = pd.read_csv("summons-15-25.csv")

# Clean nulls first
df = df.replace("(null)", np.nan)

summons_by_race = (
    df["RACE"]
    .value_counts()
    .rename_axis("Race")
    .reset_index(name="Count")
)

print(summons_by_race)
````

````python
#7. Which borough had the fewest offenses per 100K people?

# 2020 Census population
populations = {
    "BROOKLYN": 2736074,
    "QUEENS": 2405464,
    "MANHATTAN": 1694251,
    "BRONX": 1472654,
    "STATEN ISLAND": 495747
}

### Prompt: Provide the pandas code to find the borough with the fewest offenses per 100k people.


import pandas as pd
import numpy as np

df = pd.read_csv("summons-15-25.csv")

# Clean nulls
df = df.replace("(null)", np.nan)

# 2020 Census populations by borough
populations = {
    "BROOKLYN": 2736074,
    "QUEENS": 2405464,
    "MANHATTAN": 1694251,
    "BRONX": 1472654,
    "STATEN ISLAND": 495747
}

# Count summons by borough
summons_by_boro = (
    df["BORO"]
    .value_counts()
    .rename_axis("Borough")
    .reset_index(name="Count")
)

# Map population and calculate rate per 100,000
summons_by_boro["Population"] = summons_by_boro["Borough"].map(populations)
summons_by_boro["Rate_Per_100k"] = (
    summons_by_boro["Count"] / summons_by_boro["Population"] * 100_000
).round(2)

# Order by rate descending
summons_by_boro = summons_by_boro.sort_values("Rate_Per_100k", ascending=False)

print(summons_by_boro)
````
